```python
import pandas as pd
import re
from collections import Counter

# Example list of known brands (expand this as needed)
known_brands = [
    "rolex", "omega", "seiko", "cartier", "casio", "citizen", "tag heuer", "tissot", "longines",
    "panerai", "hublot", "breitling", "orient", "grand seiko", "rado", "breguet", "invicta",
    "swatch", "nixon", "movado", "apple", "bulova", "g-shock", "alpine", "audemars piguet"
]

def clean_title(title):
    title = title.lower()
    title = re.sub(r'[^\w\s]', '', title)  # remove punctuation
    title = re.sub(r'\s+', ' ', title).strip()  # remove extra spaces
    return title

def extract_model(title):
    title = clean_title(title)

    for brand in sorted(known_brands, key=len, reverse=True):
        brand_clean = brand.lower()
        if brand_clean in title:
            # Extract 2â€“4 words after the brand as model name
            parts = title.split()
            brand_index = next((i for i, word in enumerate(parts) if brand_clean in word), None)
            if brand_index is not None and brand_index + 1 < len(parts):
                model_words = parts[brand_index:brand_index+4]
                return ' '.join(model_words)

    # Fallback: use first 2-3 words of title
    return ' '.join(title.split()[:3])

# Apply extraction
df['model'] = df['title'].fillna('').apply(extract_model)

# Count model frequencies
model_counts = Counter(df['model'])

# Get top N most common models
top_models = model_counts.most_common(500)
print(top_models)
```
